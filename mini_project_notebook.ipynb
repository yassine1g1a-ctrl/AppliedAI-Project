## 5. Experimental Methodology
  
Preprocessing
- Missing and infinite values are replace by mean values
- Non-numeric features and target labels are encoded with `LabelEncoder` 
- Features are standardized with `StandardScaler` fitted on the training partition and applied to validation/test to avoid leakage.
- Constant columns and candidate ID columns (columns with a unique value per row) are removed

Validation and Splits
- Each dataset is split into train+validation (80%) and test (20%).
- Hyperparameter selection is performed by Stratified K-Fold cross-validation on the train+validation set (`n_splits=5`).
- If `StratifiedKFold` cannot be applied due to extremely small class counts, a fallback strategy assigns rare-class samples to training folds to preserve class representation during validation.

Training and Evaluation
- For every CV fold, all preprocessing steps (imputation, encoding, scaling) are fit on the training fold and then applied to the corresponding validation fold to prevent data leakage.
- After hyperparameter tuning, the selected model is retrained on the complete train+validation set and evaluated once on the held-out test set.
- Primary reported metrics are `accuracy` and `f1_macro` averaged across folds. 
- For unsupervised methods, cluster labels are mapped to class labels by majority vote on the training data before computing supervised metrics. 
- I chose to tune the models using supervised metrics (accuracy and f1-score) rather than metrics that are more commonly used for unsupervised methods. I considered this choice more relevant because it allows a direct comparison between the results of these methods and those reported in the reference article. However, this decision is debatable. An alternative approach would have been to tune the models using intrinsic clustering metrics such as the silhouette score, Davies–Bouldin index, or Calinski–Harabasz score, which better reflect the internal structure and cohesion of the clusters independently of the ground-truth labels.
- For imbalanced datasets the selection criterion favors `f1_macro` over `accuracy` when class imbalance exceeds a predefined ratio.
- For final test, we also add the metric `cohen_kappa`. For unsupervised methods, clustering quality is reported using `silhouette_score` and `davies_bouldin_score`.

Reproducibility and Implementation Details
- Random seeds are fixed (`random_state=42`) for stochastic algorithms when supported to promote reproducibility.
- All fits (preprocessing and model training) use only training data; results are averaged across folds and saved to CSV for traceability.

## Load Datasets

```python
def load_dataset(name: str) -> Tuple[pd.DataFrame, pd.Series]:
    folder = os.path.join("data", name)
    files = [f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))]
    
    # pick data file
    arff_file = f"{name}.arff"
    data_file = f"{name}.data"

    if arff_file in files:
        chosen = arff_file
        is_arff = True
    elif data_file in files:
        chosen = data_file
        is_arff = False
    else:
        raise FileNotFoundError(f"Neither '{arff_file}' nor '{data_file}' found in folder {folder}")
    path = os.path.join(folder, chosen)
    
    if is_arff:
        raw_data, meta = arff.loadarff(path)
        df = pd.DataFrame(raw_data)
        # decode if needed
        for col in df.columns:
            if df[col].dtype == object:
                df[col] = df[col].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
    else:
        if name.lower().startswith("dermatology"): # exception
            df = pd.read_csv(path, header=None, sep=",", on_bad_lines='skip')
        else:
            df = pd.read_csv(path, header=None, sep=r"\s+", on_bad_lines='skip')

    X_raw = df.iloc[:, :-1]
    y = df.iloc[:, -1].astype(str).str.strip()
    
    # remove constant columns
    X_raw = X_raw.loc[:, X_raw.nunique() > 1]

    # column ID to remove if present
    for col in X_raw.columns:
        if X_raw[col].nunique() == len(X_raw):
            X_raw = X_raw.drop(columns=[col])
            break

    X = pd.DataFrame()
    for col in X_raw.columns:
        # try convertion
        converted = pd.to_numeric(X_raw[col], errors='coerce')
        if converted.notna().mean() > 0.8:
            # numeric
            X[col] = converted
        else:
            # encoding to 0,1,2,...
            le = LabelEncoder()
            X[col] = le.fit_transform(X_raw[col].astype(str))

    return X, y

# Verification
for d in datasets:
    print(f"\nDataset : {d}")
    X, y = load_dataset(d)
    print(X.head())
    print("Labels :", y.unique()[:10])

```

## Preprocessing

```python
def preprocess(X_train: pd.DataFrame, X_test: pd.DataFrame):
    # - Impute missing values with column mean (fit on train), and apply the same fill to test
    # - fit StandardScaler on train.
    X_train = X_train.copy()
    X_test = X_test.copy()

    for col in X_train.columns:
        tr_col = pd.to_numeric(X_train[col].replace([np.inf, -np.inf], np.nan), errors='coerce')
        mean_val = tr_col.mean()
        X_train[col] = tr_col.fillna(mean_val)
        # apply same transformation to test (use train mean) because we must avoid data leakage
        X_test[col] = pd.to_numeric(X_test[col].replace([np.inf, -np.inf], np.nan), errors='coerce').fillna(mean_val)

    scaler = StandardScaler()
    # as before, we just fit scaler on train only to avoid data leakage
    scaler.fit(X_train.values)
    Xtr_scaled = pd.DataFrame(scaler.transform(X_train.values), columns=X_train.columns, index=X_train.index)
    Xt_scaled = pd.DataFrame(scaler.transform(X_test.values), columns=X_test.columns, index=X_test.index)
    return Xtr_scaled, Xt_scaled, scaler

def encode_labels(labels_train: pd.Series, labels_test: pd.Series):
    le = LabelEncoder()
    le.fit(labels_train)
    ytr = le.transform(labels_train)
    yt = le.transform(labels_test)
    return ytr, yt, le

def split_cv(X, y, n_splits=5, shuffle=True, random_state=42):
    try:
        skf = StratifiedKFold(n_splits=n_splits, shuffle=shuffle, random_state=random_state)
        folds = []
        for train_idx, test_idx in skf.split(X, y):
            folds.append((train_idx, test_idx))
        return folds

    except ValueError as e:
        print("StratifiedKFold failed:")
        pass

    #Fallback: isolate rare classes
    counts = Counter(y)
    rare_classes = {c for c, n in counts.items() if n < n_splits}

    rare_mask = np.array([c in rare_classes for c in y])
    common_mask = ~rare_mask

    idx = np.arange(len(y))
    rare_idx = idx[rare_mask]
    common_idx = idx[common_mask]

    # Stratified CV on common classes only
    skf_common = StratifiedKFold(n_splits=n_splits, shuffle=shuffle, random_state=random_state)

    folds = []
    for train_c, test_c in skf_common.split(np.zeros(len(common_idx)), y[common_mask]):
        #then we add the rare classes in the train set
        train_idx = np.concatenate([common_idx[train_c], rare_idx]) 
        test_idx = common_idx[test_c]
        folds.append((train_idx, test_idx))
    return folds
```

## Training and validation for supervised methods

```python
def train_and_evaluate_supervised(model, X_train, y_train, X_test, y_test, validation = True):
    model.fit(X_train, y_train)
    ypred = model.predict(X_test)
    acc = accuracy_score(y_test, ypred)
    f1 = f1_score(y_test, ypred, average='macro')
    if not validation:
        kappa = cohen_kappa_score(y_test, ypred)
        return {'accuracy': acc, 'f1_macro': f1, 'cohen_kappa': kappa, 'y_pred': ypred}
    return {'accuracy': acc, 'f1_macro': f1, 'y_pred': ypred}
```

```python
#example with logistic regression
X, y = load_dataset("iris")
X_tv, X_test, y_tv, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
C_values = [0.01, 0.1, 1, 10]
results = []
splits = split_cv(X_tv, y_tv, n_splits=5)

for C in C_values:
    for fold_idx, (train_idx, val_idx) in enumerate(splits):
        X_train = X_tv.iloc[train_idx]
        y_train = y_tv.iloc[train_idx]
        X_val = X_tv.iloc[val_idx]
        y_val = y_tv.iloc[val_idx]

        Xtr_s, Xval_s, scaler = preprocess(X_train, X_val)
        ytr_e, yval_e, le = encode_labels(y_train, y_val)

        model = LogisticRegression(C=C, max_iter=1000, random_state=42)
        scores = train_and_evaluate_supervised(model, Xtr_s, ytr_e, Xval_s, yval_e)

        results.append({
            "C": C,
            "fold": fold_idx,
            "accuracy": scores["accuracy"],
            "f1_macro": scores["f1_macro"]
        })

df_results = pd.DataFrame(results)
mean_scores = df_results.groupby("C").mean()
print(mean_scores)

counter = Counter(y_tv)
ratio = max(counter.values()) / min(counter.values())
if ratio > 2:
    metric_to_optimize = "f1_macro" # favor F1-score for imbalanced datasets
else:
    metric_to_optimize = "accuracy"
best_C = mean_scores[metric_to_optimize].idxmax()
print("Selected C:", best_C)

Xtr_s, Xtst_s, scaler = preprocess(X_tv, X_test)
ytr_e, ytst_e, le = encode_labels(y_tv, y_test)

model = LogisticRegression(C=best_C, max_iter=1000, random_state=42)
final_scores = train_and_evaluate_supervised(model, Xtr_s, ytr_e, Xtst_s, ytst_e, validation=False)
print("Performance on test:", final_scores)
```

```python
def train_and_evaluate_unsupervised(model, X_train, y_train, X_test, y_test, validation = True):
    model.fit(X_train)

    #assign class to clusters
    cluster_to_class = {}
    clusters_train = model.predict(X_train)
    for c in np.unique(clusters_train):
        indices = np.where(clusters_train == c)[0]
        majority_class = Counter(y_train[indices]).most_common(1)[0][0]
        cluster_to_class[c] = majority_class

    clusters_test = model.predict(X_test)
    y_pred = np.array([cluster_to_class.get(c, -1) for c in clusters_test])
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='macro')

    if not validation:
        kappa = cohen_kappa_score(y_test, y_pred)
        labels = model.predict(X_train)
        silhouette = silhouette_score(X_train, labels)
        davies_bouldin = davies_bouldin_score(X_train, labels)
        return {'accuracy': acc, 'f1_macro': f1, 'cohen_kappa': kappa, 'silhouette': silhouette, 'davies_bouldin': davies_bouldin, 'y_pred': y_pred}

    return {'accuracy': acc, 'f1_macro': f1, 'y_pred': y_pred}
```

## Models Implemented
### SOM (Unsupervised)

```python
def getEuclideanDistance(single_point, array):
    nrows, ncols, nfeatures=array.shape[0],array.shape[1], array.shape[2]
    points = array.reshape((nrows * ncols, nfeatures))
    dist = np.sqrt(np.sum((points - single_point) ** 2, axis=1))
    return dist.reshape((nrows, ncols))


class SOM:
    def __init__(self, dispRes, trainingData, ndim=10, nepochs=10, eta0=0.1, etadecay=0.05, sgm0=20, sgmdecay=0.05, showMode=0):
        self.dispRes = dispRes
        self.trainingData = trainingData
        self.ndim = ndim
        self.nepochs = nepochs
        self.eta0 = eta0
        self.etadecay = etadecay
        self.sgm0 = sgm0
        self.sgmdecay = sgmdecay
        self.showMode = showMode
        self.nfeatures = trainingData.shape[1]
        self.ntrainingvectors = trainingData.shape[0]
        self.nrows = ndim
        self.ncols = ndim
        self.som = None


    def train(self):
        mu, sigma = 0, 0.1
        np.random.seed(int(time.time()))
        self.som = np.random.normal(mu, sigma, (self.nrows, self.ncols, self.nfeatures))

        x, y = np.meshgrid(range(self.ncols), range(self.nrows))

        for t in range(1, self.nepochs+1):
            eta = self.eta0 * math.exp(-t*self.etadecay)
            sgm = self.sgm0 * math.exp(-t*self.sgmdecay)
            width = math.ceil(sgm*3)

            for ntraining in range(self.ntrainingvectors):
                trainingVector = self.trainingData[ntraining,:]
                dist = getEuclideanDistance(trainingVector, self.som)
                bmurow, bmucol = np.unravel_index(np.argmin(dist, axis=None), dist.shape)
                g = np.exp(-((np.power(x - bmucol,2)) + (np.power(y - bmurow,2))) / (2*sgm*sgm))
                fromrow = max(0,bmurow - width)
                torow   = min(bmurow + width,self.nrows)
                fromcol = max(0,bmucol - width)
                tocol   = min(bmucol + width,self.ncols)
                neighbourNeurons = self.som[fromrow:torow,fromcol:tocol,:]
                sz = neighbourNeurons.shape
                T = np.tile(trainingVector, (sz[0]*sz[1],1)).reshape((sz[0], sz[1], self.nfeatures))
                G = np.dstack([g[fromrow:torow,fromcol:tocol]]*self.nfeatures)
                neighbourNeurons = neighbourNeurons + eta * G * (T - neighbourNeurons)
                self.som[fromrow:torow,fromcol:tocol,:] = neighbourNeurons

        return self.som

    def test(self, classes, grid_):
        nclasses = np.max(classes)
        som_cl = np.zeros((self.ndim, self.ndim, nclasses+1))

        # counting classes for each neuron
        for ntraining in range(self.ntrainingvectors):
            trainingVector = self.trainingData[ntraining,:]
            class_of_sample = classes[ntraining]
            dist = getEuclideanDistance(trainingVector, self.som)
            bmurow, bmucol = np.unravel_index(np.argmin(dist, axis=None), dist.shape)
            som_cl[bmurow, bmucol, class_of_sample] += 1

        # then majority class assignment
        for i in range(self.nrows):
            for j in range(self.ncols):
                grid_[i,j] = np.argmax(som_cl[i,j,:])

        return grid_
```

### MLP (Supervised)

```python
#functions of non-linear activations
def f_sigmoid(X, deriv=False):
    if not deriv:
        return 1 / (1 + np.exp(-X))
    else:
        return f_sigmoid(X)*(1 - f_sigmoid(X))


def f_softmax(X):
    Z = np.sum(np.exp(X), axis=1)
    Z = Z.reshape(Z.shape[0], 1)
    return np.exp(X) / Z

def exit_with_err(err_str):
    print >> sys.stderr, err_str
    sys.exit(1)

#Functionality of a single hidden layer
class Layer:
    def __init__(self, size, batch_size, is_input=False, is_output=False,
                 activation=f_sigmoid):
        self.is_input = is_input
        self.is_output = is_output

        # Z is the matrix that holds output values
        self.Z = np.zeros((batch_size, size[0]))
        # The activation function is an externally defined function (with a
        # derivative) that is stored here
        self.activation = activation

        # W is the outgoing weight matrix for this layer
        self.W = None
        # S is the matrix that holds the inputs to this layer
        self.S = None
        # D is the matrix that holds the deltas for this layer
        self.D = None
        # Fp is the matrix that holds the derivatives of the activation function
        self.Fp = None

        if not is_input:
            self.S = np.zeros((batch_size, size[0]))
            self.D = np.zeros((batch_size, size[0]))

        if not is_output:
            self.W = np.random.normal(size=size, scale=1E-4)

        if not is_input and not is_output:
            self.Fp = np.zeros((size[0], batch_size))

    def forward_propagate(self):
        if self.is_input:
            return self.Z.dot(self.W)

        self.Z = self.activation(self.S)
        if self.is_output:
            return self.Z
        else:
            # For hidden layers, we add the bias values here
            self.Z = np.append(self.Z, np.ones((self.Z.shape[0], 1)), axis=1)
            self.Fp = self.activation(self.S, deriv=True).T
            return self.Z.dot(self.W)

class MultiLayerPerceptron:
    def __init__(self, layer_config, batch_size=100):
        self.layers = []
        self.num_layers = len(layer_config)
        self.minibatch_size = batch_size

        for i in range(self.num_layers-1):
            if i == 0:
                print ("Initializing input layer with size {0}.".format(layer_config[i]))
                # Here, we add an additional unit at the input for the bias
                # weight.
                self.layers.append(Layer([layer_config[i]+1, layer_config[i+1]],
                                         batch_size,
                                         is_input=True))
            else:
                print ("Initializing hidden layer with size {0}.".format(layer_config[i]))
                # Here we add an additional unit in the hidden layers for the
                # bias weight.
                self.layers.append(Layer([layer_config[i]+1, layer_config[i+1]],
                                         batch_size,
                                         activation=f_sigmoid))

        print ("Initializing output layer with size {0}.".format(layer_config[-1]))
        self.layers.append(Layer([layer_config[-1], None],
                                 batch_size,
                                 is_output=True,
                                 activation=f_softmax))
        print ("Done!")

    def forward_propagate(self, data):
        # We need to be sure to add bias values to the input
        self.layers[0].Z = np.append(data, np.ones((data.shape[0], 1)), axis=1)

        for i in range(self.num_layers-1):
            self.layers[i+1].S = self.layers[i].forward_propagate()
        return self.layers[-1].forward_propagate()

    def backpropagate(self, yhat, labels):
        self.layers[-1].D = (yhat - labels).T
        for i in range(self.num_layers-2, 0, -1):
            # We do not calculate deltas for the bias values
            W_nobias = self.layers[i].W[0:-1, :]
            
            self.layers[i].D = W_nobias.dot(self.layers[i+1].D) * self.layers[i].Fp

    def update_weights(self, eta):
        for i in range(0, self.num_layers-1):
            W_grad = -eta*(self.layers[i+1].D.dot(self.layers[i].Z)).T
            self.layers[i].W += W_grad

    def evaluate(self, train_data, train_labels, test_data, test_labels,
                 num_epochs=70, eta=0.05, eval_train=False, eval_test=True):

        N_train = len(train_labels)*len(train_labels[0])
        N_test = len(test_labels)*len(test_labels[0])

        print ("Training for {0} epochs...".format(num_epochs))
        for t in range(0, num_epochs):
            out_str = "[{0:4d}] ".format(t)

            for b_data, b_labels in zip(train_data, train_labels):
                output = self.forward_propagate(b_data)
                self.backpropagate(output, b_labels)

                self.update_weights(eta=eta)

            if eval_train:
                errs = 0
                for b_data, b_labels in zip(train_data, train_labels):
                    output = self.forward_propagate(b_data)
                    yhat = np.argmax(output, axis=1)
                    errs += np.sum(1-b_labels[np.arange(len(b_labels)), yhat])

                out_str = ("{0} Training error: {1:.5f}".format(out_str,
                                                           float(errs)/N_train))

            if eval_test:
                errs = 0
                for b_data, b_labels in zip(test_data, test_labels):
                    output = self.forward_propagate(b_data)
                    yhat = np.argmax(output, axis=1)
                    errs += np.sum(1-b_labels[np.arange(len(b_labels)), yhat])

                out_str = ("{0} Test error: {1:.5f}").format(out_str,
                                                       float(errs)/N_test)

            print (out_str)
def label_to_bit_vector(labels, nbits):
    bit_vector = np.zeros((labels.shape[0], nbits))
    for i in range(labels.shape[0]):
        bit_vector[i, labels[i]] = 1.0

    return bit_vector

def create_batches(data, labels, batch_size, nb_labels, create_bit_vector=False):
    N = data.shape[0]
    print ("Batch size {0}, the number of examples {1}.".format(batch_size,N))

    chunked_data = []
    chunked_labels = []
    idx = 0
    while idx < N:
        chunked_data.append(data[idx:idx+batch_size, :])
        if not create_bit_vector:
            chunked_labels.append(labels[idx:idx+batch_size])
        else:
            bit_vector = label_to_bit_vector(labels[idx:idx+batch_size], nb_labels)
            chunked_labels.append(bit_vector)

        idx += batch_size

    return chunked_data, chunked_labels

def prepare_for_backprop(batch_size, Train_images, Train_labels, Valid_images, Valid_labels):
    
    print ("Creating data...")
    all_labels = np.concatenate([Train_labels, Valid_labels])
    nb_labels = len(np.unique(all_labels))
    batched_train_data, batched_train_labels = create_batches(Train_images, Train_labels,
                                              batch_size,
                                              nb_labels,
                                              create_bit_vector=True)
    batched_valid_data, batched_valid_labels = create_batches(Valid_images, Valid_labels,
                                              batch_size,
                                              nb_labels,
                                              create_bit_vector=True)
    print ("Done!")


    return batched_train_data, batched_train_labels,  batched_valid_data, batched_valid_labels

```

```python
def tune_logistic_regression(X_tv, y_tv, X_test, y_test):
    C_values = [0.01, 0.1, 1, 10]
    results = []
    splits = split_cv(X_tv, y_tv, n_splits=5)
    
    for C in C_values:
        for fold_idx, (train_idx, val_idx) in enumerate(splits):
            X_train = X_tv.iloc[train_idx]
            y_train = y_tv.iloc[train_idx]
            X_val = X_tv.iloc[val_idx]
            y_val = y_tv.iloc[val_idx]
            
            Xtr_s, Xval_s, scaler = preprocess(X_train, X_val)
            ytr_e, yval_e, le = encode_labels(y_train, y_val)
            
            model = LogisticRegression(C=C, max_iter=1000, random_state=42)
            scores = train_and_evaluate_supervised(model, Xtr_s, ytr_e, Xval_s, yval_e)
            
            results.append({
                "C": C,
                "fold": fold_idx,
                "accuracy": scores["accuracy"],
                "f1_macro": scores["f1_macro"]
            })
    
    df_results = pd.DataFrame(results)
    mean_scores = df_results.groupby("C").mean()
    
    counter = Counter(y_tv)
    ratio = max(counter.values()) / min(counter.values())
    metric_to_optimize = "f1_macro" if ratio > 2 else "accuracy"
    best_C = mean_scores[metric_to_optimize].idxmax()

    Xtr_s, Xtst_s, scaler = preprocess(X_tv, X_test)
    ytr_e, ytst_e, le = encode_labels(y_tv, y_test)
    
    model = LogisticRegression(C=best_C, max_iter=1000, random_state=42)
    final_scores = train_and_evaluate_supervised(model, Xtr_s, ytr_e, Xtst_s, ytst_e, validation=False)
    
    return final_scores, best_C


def tune_random_forest(X_tv, y_tv, X_test, y_test):
    n_estimators_values = [100, 300]
    max_depth_values = [None, 10]
    results = []
    splits = split_cv(X_tv, y_tv, n_splits=5)
    
    for n_est in n_estimators_values:
        for max_d in max_depth_values:
            for fold_idx, (train_idx, val_idx) in enumerate(splits):
                X_train = X_tv.iloc[train_idx]
                y_train = y_tv.iloc[train_idx]
                X_val = X_tv.iloc[val_idx]
                y_val = y_tv.iloc[val_idx]
                
                Xtr_s, Xval_s, scaler = preprocess(X_train, X_val)
                ytr_e, yval_e, le = encode_labels(y_train, y_val)
                
                model = RandomForestClassifier(n_estimators=n_est, max_depth=max_d, random_state=42)
                scores = train_and_evaluate_supervised(model, Xtr_s, ytr_e, Xval_s, yval_e)
                
                results.append({
                    "n_estimators": n_est,
                    "max_depth": max_d,
                    "fold": fold_idx,
                    "accuracy": scores["accuracy"],
                    "f1_macro": scores["f1_macro"]
                })
    
    df_results = pd.DataFrame(results)
    mean_scores = df_results.groupby(["n_estimators", "max_depth"]).mean()
    
    counter = Counter(y_tv)
    ratio = max(counter.values()) / min(counter.values())
    metric_to_optimize = "f1_macro" if ratio > 2 else "accuracy"
    best_params = mean_scores[metric_to_optimize].idxmax()
    best_n_est, best_max_d = best_params

    best_n_est = int(best_n_est)
    best_max_d = None if pd.isna(best_max_d) else int(best_max_d)
    
    # Train final model
    Xtr_s, Xtst_s, scaler = preprocess(X_tv, X_test)
    ytr_e, ytst_e, le = encode_labels(y_tv, y_test)
    
    model = RandomForestClassifier(n_estimators=best_n_est, max_depth=best_max_d, random_state=42)
    final_scores = train_and_evaluate_supervised(model, Xtr_s, ytr_e, Xtst_s, ytst_e, validation=False)
    
    return final_scores, (best_n_est, best_max_d)


def tune_mlp(X_tv, y_tv, X_test, y_test):
    hidden_layers_values = [50, 100]
    learning_rate_values = [0.001, 0.01]
    results = []
    splits = split_cv(X_tv, y_tv, n_splits=5)
    
    for hidden_size in hidden_layers_values:
        for lr in learning_rate_values:
            for fold_idx, (train_idx, val_idx) in enumerate(splits):
                X_train = X_tv.iloc[train_idx]
                y_train = y_tv.iloc[train_idx]
                X_val = X_tv.iloc[val_idx]
                y_val = y_tv.iloc[val_idx]
                
                Xtr_s, Xval_s, scaler = preprocess(X_train, X_val)
                ytr_e, yval_e, le = encode_labels(y_train, y_val)
                
                # MLP
                n_classes = len(np.unique(ytr_e))
                n_features = Xtr_s.shape[1]
                n_train = Xtr_s.shape[0]
                batch_size = max([b for b in range(1, 33) if n_train % b == 0])
                
                bt_train, bl_train, bt_val, bl_val = prepare_for_backprop(
                    batch_size, Xtr_s.values, ytr_e, Xval_s.values, yval_e
                )
                
                layer_config = [n_features, hidden_size, n_classes]
                mlp = MultiLayerPerceptron(layer_config, batch_size=batch_size)
                mlp.evaluate(bt_train, bl_train, bt_val, bl_val, num_epochs=50,
                           eta=lr, eval_train=False, eval_test=False)
                
                outputs = mlp.forward_propagate(Xval_s.values)
                y_pred = np.argmax(outputs, axis=1)
                acc = accuracy_score(yval_e, y_pred)
                f1 = f1_score(yval_e, y_pred, average='macro')
                
                results.append({
                    "hidden_size": hidden_size,
                    "learning_rate": lr,
                    "fold": fold_idx,
                    "accuracy": acc,
                    "f1_macro": f1
                })
    
    df_results = pd.DataFrame(results)
    mean_scores = df_results.groupby(["hidden_size", "learning_rate"]).mean()
    
    counter = Counter(y_tv)
    ratio = max(counter.values()) / min(counter.values())
    metric_to_optimize = "f1_macro" if ratio > 2 else "accuracy"
    best_params = mean_scores[metric_to_optimize].idxmax()
    best_hidden, best_lr = best_params
    
    # Train final model
    Xtr_s, Xtst_s, scaler = preprocess(X_tv, X_test)
    ytr_e, ytst_e, le = encode_labels(y_tv, y_test)
    
    n_classes = len(np.unique(ytr_e))
    n_features = Xtr_s.shape[1]
    n_train = Xtr_s.shape[0]
    batch_size = max([b for b in range(1, 33) if n_train % b == 0])
    
    bt_train, bl_train, bt_test, bl_test = prepare_for_backprop(
        batch_size, Xtr_s.values, ytr_e, Xtst_s.values, ytst_e
    )
    
    layer_config = [n_features, best_hidden, n_classes]
    mlp = MultiLayerPerceptron(layer_config, batch_size=batch_size)
    mlp.evaluate(bt_train, bl_train, bt_test, bl_test, num_epochs=50,
               eta=best_lr, eval_train=False, eval_test=False)
    
    outputs = mlp.forward_propagate(Xtst_s.values)
    y_pred = np.argmax(outputs, axis=1)
    acc = accuracy_score(ytst_e, y_pred)
    f1 = f1_score(ytst_e, y_pred, average='macro')
    kappa = cohen_kappa_score(ytst_e, y_pred)
    
    final_scores = {'accuracy': acc, 'f1_macro': f1, 'cohen_kappa': kappa, 'y_pred': y_pred}
    
    return final_scores, (best_hidden, best_lr)


def tune_kmeans(X_tv, y_tv, X_test, y_test):
    n_clusters_values = [2, 3, 4]
    results = []
    splits = split_cv(X_tv, y_tv, n_splits=5)
    
    for n_clust in n_clusters_values:
        for fold_idx, (train_idx, val_idx) in enumerate(splits):
            X_train = X_tv.iloc[train_idx]
            y_train = y_tv.iloc[train_idx]
            X_val = X_tv.iloc[val_idx]
            y_val = y_tv.iloc[val_idx]
            
            Xtr_s, Xval_s, scaler = preprocess(X_train, X_val)
            ytr_e, yval_e, le = encode_labels(y_train, y_val)
            
            model = KMeans(n_clusters=n_clust, random_state=42, n_init=10)
            scores = train_and_evaluate_unsupervised(model, Xtr_s, ytr_e, Xval_s, yval_e)
            
            results.append({
                "n_clusters": n_clust,
                "fold": fold_idx,
                "accuracy": scores["accuracy"],
                "f1_macro": scores["f1_macro"]
            })
    
    df_results = pd.DataFrame(results)
    mean_scores = df_results.groupby("n_clusters").mean()
    
    counter = Counter(y_tv)
    ratio = max(counter.values()) / min(counter.values())
    metric_to_optimize = "f1_macro" if ratio > 1.5 else "accuracy"
    best_n_clust = mean_scores[metric_to_optimize].idxmax()
    
    # Train final model
    Xtr_s, Xtst_s, scaler = preprocess(X_tv, X_test)
    ytr_e, ytst_e, le = encode_labels(y_tv, y_test)
    
    model = KMeans(n_clusters=best_n_clust, random_state=42, n_init=10)
    final_scores = train_and_evaluate_unsupervised(model, Xtr_s, ytr_e, Xtst_s, ytst_e, validation=False)
    
    return final_scores, best_n_clust


def tune_gmm(X_tv, y_tv, X_test, y_test):
    n_components_values = [2, 3, 4, 5]
    results = []
    splits = split_cv(X_tv, y_tv, n_splits=5)
    
    for n_comp in n_components_values:
        for fold_idx, (train_idx, val_idx) in enumerate(splits):
            X_train = X_tv.iloc[train_idx]
            y_train = y_tv.iloc[train_idx]
            X_val = X_tv.iloc[val_idx]
            y_val = y_tv.iloc[val_idx]
            
            Xtr_s, Xval_s, scaler = preprocess(X_train, X_val)
            ytr_e, yval_e, le = encode_labels(y_train, y_val)
            
            model = GaussianMixture(n_components=n_comp, random_state=42)
            scores = train_and_evaluate_unsupervised(model, Xtr_s, ytr_e, Xval_s, yval_e)
            
            results.append({
                "n_components": n_comp,
                "fold": fold_idx,
                "accuracy": scores["accuracy"],
                "f1_macro": scores["f1_macro"]
            })
    
    df_results = pd.DataFrame(results)
    mean_scores = df_results.groupby("n_components").mean()
    
    counter = Counter(y_tv)
    ratio = max(counter.values()) / min(counter.values())
    metric_to_optimize = "f1_macro" if ratio > 1.5 else "accuracy"
    best_n_comp = mean_scores[metric_to_optimize].idxmax()
    
    # Train final model
    Xtr_s, Xtst_s, scaler = preprocess(X_tv, X_test)
    ytr_e, ytst_e, le = encode_labels(y_tv, y_test)
    
    model = GaussianMixture(n_components=best_n_comp, random_state=42)
    final_scores = train_and_evaluate_unsupervised(model, Xtr_s, ytr_e, Xtst_s, ytst_e, validation=False)
    
    return final_scores, best_n_comp


def tune_som(X_tv, y_tv, X_test, y_test):
    grid_values = [5, 10]
    learning_rate_values = [0.1, 0.5]
    results = []
    splits = split_cv(X_tv, y_tv, n_splits=5)
    
    for grid_size in grid_values:
        for eta in learning_rate_values:
            for fold_idx, (train_idx, val_idx) in enumerate(splits):
                X_train = X_tv.iloc[train_idx]
                y_train = y_tv.iloc[train_idx]
                X_val = X_tv.iloc[val_idx]
                y_val = y_tv.iloc[val_idx]
                
                Xtr_s, Xval_s, scaler = preprocess(X_train, X_val)
                ytr_e, yval_e, le = encode_labels(y_train, y_val)
                
                # Train SOM
                som = SOM(dispRes=None, trainingData=Xtr_s.values, ndim=grid_size, 
                         nepochs=10, eta0=eta)
                som.train()
                grid = np.zeros((som.nrows, som.ncols), dtype=int)
                som.test(ytr_e, grid)
                
                # Predict on validation
                def som_predict(som_obj, Xdf, grid_map):
                    preds = []
                    for p in Xdf.values:
                        dist = getEuclideanDistance(p, som_obj.som)
                        r, c = np.unravel_index(np.argmin(dist), dist.shape)
                        preds.append(int(grid_map[r, c]))
                    return np.array(preds)
                
                y_pred = som_predict(som, Xval_s, grid)
                acc = accuracy_score(yval_e, y_pred)
                f1 = f1_score(yval_e, y_pred, average='macro')
                
                results.append({
                    "grid_size": grid_size,
                    "learning_rate": eta,
                    "fold": fold_idx,
                    "accuracy": acc,
                    "f1_macro": f1
                })
    
    df_results = pd.DataFrame(results)
    mean_scores = df_results.groupby(["grid_size", "learning_rate"]).mean()
    
    counter = Counter(y_tv)
    ratio = max(counter.values()) / min(counter.values())
    metric_to_optimize = "f1_macro" if ratio > 1.5 else "accuracy"
    best_params = mean_scores[metric_to_optimize].idxmax()
    best_grid, best_eta = best_params
    
    # Train final model
    Xtr_s, Xtst_s, scaler = preprocess(X_tv, X_test)
    ytr_e, ytst_e, le = encode_labels(y_tv, y_test)
    
    som = SOM(dispRes=None, trainingData=Xtr_s.values, ndim=best_grid, 
             nepochs=10, eta0=best_eta)
    som.train()
    grid = np.zeros((som.nrows, som.ncols), dtype=int)
    som.test(ytr_e, grid)
    
    def som_predict(som_obj, Xdf, grid_map):
        preds = []
        for p in Xdf.values:
            dist = getEuclideanDistance(p, som_obj.som)
            r, c = np.unravel_index(np.argmin(dist), dist.shape)
            preds.append(int(grid_map[r, c]))
        return np.array(preds)
    
    y_pred = som_predict(som, Xtst_s, grid)
    acc = accuracy_score(ytst_e, y_pred)
    f1 = f1_score(ytst_e, y_pred, average='macro')
    kappa = cohen_kappa_score(ytst_e, y_pred)
    
    # Compute clustering metrics on training data
    train_bmu = []
    for p in Xtr_s.values:
        dist = getEuclideanDistance(p, som.som)
        r, c = np.unravel_index(np.argmin(dist), dist.shape)
        train_bmu.append(r * som.ncols + c)
    train_bmu = np.array(train_bmu)
    
    silhouette = silhouette_score(Xtr_s.values, train_bmu)
    davies_bouldin = davies_bouldin_score(Xtr_s.values, train_bmu)
    
    final_scores = {'accuracy': acc, 'f1_macro': f1, 'cohen_kappa': kappa, 
                   'silhouette': silhouette, 'davies_bouldin': davies_bouldin, 'y_pred': y_pred}
    
    return final_scores, (best_grid, best_eta)
```

## Main Evaluation Loop with Hyperparameter Tuning

```python
EVAL_DATASETS = ['iris', 'wine', 'waveform', 'car', 'titanic', 'ionosphere', 
                 'dermatology', 'ecoli', 'zoo', 'page-blocks', 'spambase', 
                 'acute-inflammation', 'vertebral-column-3clases', 
                 'vertebral-column-2clases', 'tic-tac-toe', 'synthetic-control', 
                 'statlog-vehicle', 'statlog-heart', 'seeds', 'balance-scale', 
                 'acute-nephritis', 'heart-hungarian']
all_results = []
best_hyperparams = []

for ds_name in EVAL_DATASETS:
    print(f"dataset: {ds_name}")
    try:
        X, y = load_dataset(ds_name)
        # Split train+validation (80%) and test (20%)
        X_tv, X_test, y_tv, y_test = train_test_split(X, y, test_size=0.2, 
                                                      stratify=y, 
                                                      random_state=42)
        
        # LogisticRegression
        print("\nTuning LogisticRegression...")
        lr_scores, lr_params = tune_logistic_regression(X_tv, y_tv, X_test, y_test)
        all_results.append({
            'dataset': ds_name,
            'model': 'LogisticRegression',
            'accuracy': lr_scores['accuracy'],
            'f1_macro': lr_scores['f1_macro'],
            'cohen_kappa': lr_scores['cohen_kappa']
        })
        best_hyperparams.append({
            'dataset': ds_name,
            'model': 'LogisticRegression',
            'hyperparams': f"C={lr_params}"
        })
        
        # RandomForest
        print("Tuning RandomForest...")
        rf_scores, rf_params = tune_random_forest(X_tv, y_tv, X_test, y_test)
        all_results.append({
            'dataset': ds_name,
            'model': 'RandomForest',
            'accuracy': rf_scores['accuracy'],
            'f1_macro': rf_scores['f1_macro'],
            'cohen_kappa': rf_scores['cohen_kappa']
        })
        best_hyperparams.append({
            'dataset': ds_name,
            'model': 'RandomForest',
            'hyperparams': f"n_estimators={rf_params[0]}, max_depth={rf_params[1]}"
        })
        
        # MLP
        print("Tuning MLP...")
        mlp_scores, mlp_params = tune_mlp(X_tv, y_tv, X_test, y_test)
        all_results.append({
            'dataset': ds_name,
            'model': 'MLP',
            'accuracy': mlp_scores['accuracy'],
            'f1_macro': mlp_scores['f1_macro'],
            'cohen_kappa': mlp_scores['cohen_kappa']
        })
        best_hyperparams.append({
            'dataset': ds_name,
            'model': 'MLP',
            'hyperparams': f"hidden_size={mlp_params[0]}, lr={mlp_params[1]}"
        })
        
        # KMeans
        print("Tuning KMeans...")
        km_scores, km_params = tune_kmeans(X_tv, y_tv, X_test, y_test)
        all_results.append({
            'dataset': ds_name,
            'model': 'KMeans',
            'accuracy': km_scores['accuracy'],
            'f1_macro': km_scores['f1_macro'],
            'cohen_kappa': km_scores['cohen_kappa'],
            'silhouette': km_scores['silhouette'],
            'davies_bouldin': km_scores['davies_bouldin']
        })
        best_hyperparams.append({
            'dataset': ds_name,
            'model': 'KMeans',
            'hyperparams': f"n_clusters={km_params}"
        })
        
        # GMM
        print("Tuning GMM...")
        gmm_scores, gmm_params = tune_gmm(X_tv, y_tv, X_test, y_test)
        all_results.append({
            'dataset': ds_name,
            'model': 'GMM',
            'accuracy': gmm_scores['accuracy'],
            'f1_macro': gmm_scores['f1_macro'],
            'cohen_kappa': gmm_scores['cohen_kappa'],
            'silhouette': gmm_scores['silhouette'],
            'davies_bouldin': gmm_scores['davies_bouldin']
        })
        best_hyperparams.append({
            'dataset': ds_name,
            'model': 'GMM',
            'hyperparams': f"n_components={gmm_params}"
        })
        
        # SOM
        print("Tuning SOM...")
        som_scores, som_params = tune_som(X_tv, y_tv, X_test, y_test)
        all_results.append({
            'dataset': ds_name,
            'model': 'SOM',
            'accuracy': som_scores['accuracy'],
            'f1_macro': som_scores['f1_macro'],
            'cohen_kappa': som_scores['cohen_kappa'],
            'silhouette': som_scores['silhouette'],
            'davies_bouldin': som_scores['davies_bouldin']
        })
        best_hyperparams.append({
            'dataset': ds_name,
            'model': 'SOM',
            'hyperparams': f"grid_size={som_params[0]}, eta0={som_params[1]}"
        })
        
        print(f"\nCompleted dataset: {ds_name}")
        
    except Exception as e:
        print(f"\nError for dataset {ds_name}: {str(e)}")
        continue


df_new_results = pd.DataFrame(all_results)
df_new_hyperparams = pd.DataFrame(best_hyperparams)

# Merge with exsting results if file exists (avoid overwriting, except duplicates)
results_path = 'final_results_with_tuning.csv'
hyperparams_path = 'best_hyperparameters.csv'

if os.path.isfile(results_path):
    df_old_results = pd.read_csv(results_path)
    df_results = pd.concat([df_old_results, df_new_results], ignore_index=True)
    df_results = df_results.drop_duplicates(subset=['dataset', 'model'], keep='last')
else:
    df_results = df_new_results

if os.path.isfile(hyperparams_path):
    df_old_hyperparams = pd.read_csv(hyperparams_path)
    df_hyperparams = pd.concat([df_old_hyperparams, df_new_hyperparams], ignore_index=True)
    df_hyperparams = df_hyperparams.drop_duplicates(subset=['dataset', 'model'], keep='last')
else:
    df_hyperparams = df_new_hyperparams

# Save
df_results.to_csv(results_path, index=False)
df_hyperparams.to_csv(hyperparams_path, index=False)

# summary statistics
print("Perf by model:")
summary = df_results.groupby('model')[['accuracy', 'f1_macro', 'cohen_kappa']].mean()
print(summary.round(4))
```

### Discussion

- Supervised models (Random Forest, Logistic Regression, MLP) generally outperform unsupervised models in datasets with well-defined class labels.  
- Clustering algorithms are effective only on datasets with clear cluster separation (e.g., wine, acute-inflammation, iris).  
- SOM is less reliable for clustering in datasets with high noise or overlapping classes.  
- Dataset characteristics (number of classes, feature dimensionality, and separability) significantly impact model performance.
